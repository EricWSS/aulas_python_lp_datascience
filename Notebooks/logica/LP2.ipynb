## Sintaxe das condicionais em *Python*

### *if*
Para verificarmos alguma condição começamos sempre pelo comando *if*, essa estrutura permite que o programa execute um determinado bloco de código, somente se a condição passada seja veridadeira.

Exemplo de código utilizando condicionais:

```python
nome = 'Felipe'

if nome == 'Felipe':
  print(f'Seu nome é {nome}')
```

Neste caso, estamos verificando se a variável nome possui o valor 'Felipe' atribuida a ela, se for verdadeira, o comando *print()* vai acontecer, se não, nada irá acontecer.

Se você estiver atento o *print()* está mais para o lado direito, isso se chama indentação, toda vez que abrirmos um bloco de condição precisamos utilizar essa indentação para informar ao programa que aqueles comandos só irão acontecer se a condição for verdadeira, ou seja, esse é nosso bloco de código.

```python
# Dessa maneira, podemos passar algo depois do nosso bloco de código informando
# que isso é independente da condicional, exemplo:

numero = 1

if numero == 1:
  print('O número é igual a 1')

print('Esse print vai acontecer independente da condicional ser verdadeira')
```

```python
numero = 1

if numero != 1:
  print('O número é igual a 1')

print('Esse print vai acontecer independente da condicional ser verdadeira')
```

Certo, mas se quisermos verificar novas condições?

Nesse caso, o *Python* tem um comando chamado *elif*, ele será como a condicional *if*, porém, se o *if* não for verdadeiro, ele vai verificar o *elif* e caso o *elif* seja verdadeiro ele irá executar todo o bloco do *elif*.

Vamos a um exemplo do *elif*:

```python
numero1 = 100
numero2 = 200

# Primeiro fazemos a verificação com o if
if numero1 == numero2:
  print('Os números são iguais')
elif numero1 != numero2:
  print('Os números são diferentes')
```

Nesse caso, o *if* ficou responsável por verificar se as variáveis possuem o mesmo valor e o *elif* ficou responsável por verificar se as variáveis são diferentes.

Nesse caso, se quisermos criar várias condições vamos utilizar o comando *elif*.

Por exemplo:

```python
nome = 'Felipe'

if nome == 'Maria':
  print('O nome é Maria')
elif nome == 'Adriano':
  print('O nome é Adriano')
elif nome == 'Hardmann':
  print('O nome é Hardmann')
elif nome == 'Felipe':
  print('O nome é Felipe')
```

Certo, mas, se nenhuma condição for verdadeira, o que vai acontecer com nosso programa? Simplesmente não vai apresentar nada?

**NÃO**, no *Python* nós temos mais uma condição, que é o *else*, a diferença dele para o *if* e *elif* é que ele não vai precisar de condição, ele será executado se todas as condicionais forem *false*, então vamos ver a sua sintaxa utilizando o mesmo exemplo acima.

```python
nome = 'Felipe'

if nome == 'Maria':
  print('O nome é Maria')
elif nome == 'Adriano':
  print('O nome é Adriano')
elif nome == 'Hardmann':
  print('O nome é Hardmann')
elif nome == 'Fernando':
  print('O nome é Fernando')
else:
  '''
  Perceba que o else não precisa de verificação e que nenhum nome é igual nas condicionais.
  '''
  print('Eu sou o else')
```

Certo, agora que vimos as três condicionais, *if*, *elif* e *else*, vamos praticar, lembre-se, se podemos verificar variáveis "fixas", podemos verificar variáveis que são digitadas pelo usuário. ### Outras verificações com condicionais

Em Python nós também a nossa palavra reservada *in*, que vai percorrer, normalmente, strings.

```python
escola = 'infinity'

if 'i' in escola:
  print('Há um i dentro dessa variável')
```

Dessa maneira, ele percorre toda essa estrutura de texto e verifica se há algum i dentro dessa estrutura, se houver, ele retornará um *True* e a condição será verdadeira.

Certo, outra dúvida muito frequente é como fazemos mais de uma verificação em uma condicional, como por exemplo, se eu quiser verificar a altura da pessoa e o seu peso, como eu faria?

Nesse caso temos o operador lógico chamado *and*.  

```python
altura = 1.89
peso = 76

if altura > 1.80 and peso > 76:
  print('Você está com o peso ideal')
```

No caso do *and*, as duas operações precisam ser verdadeiras para que a condição toda seja do tipo *True*.

Dessa forma, podemos fazer mais um execício:

- Construa um programa que peça ao usuário 3 números, depois retorne qual é o maior entre eles

```python
# Questão resolvida
num1 = int(input('Digite o 1° número: '))
num2 = int(input('Digite o 2° número: '))
num3 = int(input('Digite o 3° número: '))

if num1 > num2 and num1 > num3:
  print(f'O maior número é {num1}')
elif num2 > num1 and num2 > num3:
  print(f'O maior número é {num2}')
else:
  print(f'O maior número é {num3}')
```

Outro operador lógico que temos em *Python* é o *or*, esse operador vai receber mais de uma condição porém irá retornar *True* se apenas uma já for verdadeira.

Por exemplo:

```python
# Vamos verificar a questão de comparação do número 1 com o número 2
# Supondo que saibamos que só podem haver no máximo dois números iguais, vamos verificar se número 1 é menor do que número 2
num1 = 5
num2 = 10
num3 = 15

if num1 < num2 or num2 == num3:
  print(f'O número {num1} é menor do que {num2}')
```

Perceba que o número 2 não é igual ao número 3, mas só de haver uma condição verdadeira a condição por completa é verdadeira.

Com isso, vamos fazer mais uma questão.
