#O que é a Ciência de Dados?

Definição de mercado: é o processo de exploração, manipulação e análise dos dados para a descoberta e previsão através da criação de hipóteses, testes e validação com o objetivo de responder e/ou fazer recomendações capazes de serem diferenciais de negócio.

Todo esse processo precisa ter um forte embasamento estatístico e matemático e ser diretamente ligado ao negócio, além de poder ser feito de forma escalável e replicável.

Certo, algo muito importante que um cientista de dados deve sempre fazer é otimizar o uso de memória, já que, procuramos sempre manter o nosso código o mais rápido possível, pois, quando utilizarmos uma base de dados maior, se nossos dados estiverem bem organizados, podemos otimizar o uso da memória do computador.

**OBS**: Nossos tratamentos de dados ficam armazenados em memória.

Na coluna *preço garrafa* está como **float64**, que ocupa bastante espaço de memória.

A biblioteca *Pandas* nos ajuda a modificar esse espaço de memória e até mesmo o tipo de alguns dados, com a função *to_numeric()*.

Com este método, estamos fazendo algo chamado de downcast, que vai otimizar nossa memória.

**Dica**
- use o *integer* para as colunas com os valores numéricos
- use float para colunas com valores numéricos com ponto flutuante

Bem parecido com o SQL

##Alterando o tipo de dado de uma coluna numérica

```python
df['preço da garrafa'] = pd.to_numeric(df['preço da garrafa'], downcast='float')
```

Após o uso do método to_numeric, chamamos novamente o método info e percebemos que, agora, a coluna preço da garrafa está como float32, um tipo de dados mais adequado para os valores desta coluna.

OBS: com isso diminuímos o uso de memória do dataframe em questão

```python
df.info()
```

Dessa forma, agora que conhecemos o método para Downcast, tente fazer esse método na coluna *quantidade de garrafas*

```python
df['quantidade de garrafas'] = pd.to_numeric(df['quantidade de garrafas'], downcast='integer')
```

```python
df.info()
```

###Outra forma de conhecermos melhor nosso dataset é com:
####describe()

Esse método, irá retornar pra gente um resumo sobre as Series numéricas com várias informações:


  **count** : mostra quantos registros / linhas há em nosso dataframe         
  **mean** : mostra a média. Vale lembrar que a depender da coluna, esse valor pode não fazer muito sentido mas no nosso exemplo, a média se aplica as colunas de quantidade de garrafas, preço da garrafa e total.    
  **std** : desvio padrão, esse tópico é abordado na especialização em Data Science                                                              
  **min** : valor mínimo                                                  
  os **quartis**:                                                             
        **25%** : 1º quadrante, significa que 25% dos dados estão abaixo de determinada quantidade. Ex: Olhando para a coluna de quantidade de garrafas, temos o valor de 5. Isso significa dizer que 25% dos registros do nosso dataframe estão abaixo de 5 garrafas.                            
        **50%** : 2º quadrante, também conhecido como mediana. Metade dos registros do nosso dataframe estão abaixo de 8 garrafas, ainda olhando para a coluna de quantidade de garrafas.                                  
        **75%** : 3º quadrante, onde temos 75% dos registros do nosso dataframe abaixo de 10 garrafas.
  max : valor máximo                                                      
 

```python
df.describe()
```

#E o que significa esses atributos que visualizamos?


- **count** : retorna quantas linhas que há no nosso dataframe
- **unique** : retorna os valores únicos encontrados nesta Serie específica
- **top** : retorno o "termo" encontrado com maior frequência. Vale ressaltar que esta informação retornará a primeira ocorrência, caso haja, mais termos com a mesma frequência.
- **freq** : retorna o número de ocorrências do no termo listados em **top**
