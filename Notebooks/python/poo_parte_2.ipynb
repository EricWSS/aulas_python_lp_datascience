###Objeto

São instâncias das classes, a parte concreta onde armazenará os dados verdadeiros.

Esses foram os conceitos iniciais que vimos na aula passada, agora, veremos outro conceito, **herança**.

###Herança

De forma rápida, a herança permite basear uma nova classe na definição de um outra classe previamente existente.

Essas classes são criadas para facilitar o processo de estruturação, já que vamos usar uma classe "Pai" para basearmos todas as outras.

Normalmente, essa classe NÃO deve ser acessada de forma direta.

###SuperClasses

Chamamos essas classes Pai de superClasses, essas classes terão 1 ou vários descendentes. Nesse caso, podemos utilizar o exemplo de Animal.

Captura de tela de 2023-09-23 12-06-01.png

Nesse exemplo acima, a classe Animal nunca será instânciada de forma direta, mas sim suas classes filhas, que nesse caso são Mamífero e Ave.

É importante destacar que uma classe filha
HERDA todos os atributos e métodos da
classe pai, ou seja, tanto a classe Mamífero
quanto a classe Ave terão os atributos e
métodos da classe Animal.

Vamos testar esse exemplo acima:

```python
class Animal:
  def __init__(self, peso: float, idade: int, membros: int) -> None:
    self.peso = peso
    self.idade = idade
    self.membros = membros

  def locomover(self) -> str:
    return 'Animal se locomovendo'

  def alimentar(self) -> str:
    return 'Animal se alimentando'

  def emitir_som(self) -> str:
    return 'Animal emitindo som'
```

```python
class Mamifero(Animal):
  def __init__(self, tem_pelo: bool = True, cor: str = 'caramelo') -> None:
    self.pelo = tem_pelo
    self.cor = cor if tem_pelo else None
```

```python
cachorro = Mamifero(True)
print(cachorro.__dict__)
print(
    cachorro.alimentar(),
    cachorro.locomover(),
    cachorro.emitir_som(),
    sep='\n'
)
```

Perceba que não temos todos os atributos da Classe Animal na classe Mamifero, nesse caso, vamos utilizar uma função especial.

###Super()

É uma função que chama uma funcionalidade da super classe (classe pai ou classe mãe).

Vamos refatorar a classe Mamifero e mudar seu construtor para que possamos acessar o método construtor da super classe.

```python
class Mamifero(Animal):
  def __init__(self, peso: float, idade: int, membros: int,
               tem_pelo: bool = True, cor: str = 'caramelo') -> None:
    super().__init__(peso, idade, membros)
    self.pelo = tem_pelo
    self.cor = cor if tem_pelo else None
```

Dessa vez, estamos chamando os atributos de Animal, que nesse caso é peso, idade e membros.

Vamos testar se passarmos apenas os atributos da classe.

```python
gato = Mamifero(tem_pelo=True, cor='Preto')
```

Agora, nossa classe pede para que haja todos aqueles atributos, peso, idade e membros.

```python
cachorro = Mamifero(10, 3, 4, True, 'Preto')
print(cachorro.__dict__)
print(
    cachorro.alimentar(),
    cachorro.locomover(),
    cachorro.emitir_som(),
    sep='\n'
)
```

Agora sim, nosso objeto tem todos os atributos tanto da Classe Animal, quanto da classe Mamifero.
