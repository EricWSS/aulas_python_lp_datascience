### Fluxogramas
#### Primeiro, precisamos entender o conceito de fluxograma.

Um fluxograma é um diagrama que descreve, através de símbolos, um processo, sistema ou algoritmo de computador. Esse processo não é usado apenas na área de programação, podemos usar esse fluxo para estudos, planejamentos, comunicar processos complexos e diversas outras coisas.

### Certo e por que um simples fluxograma pode ser um diferencial para melhorar a sua lógica de programação e acrescentar muito na sua carreira?

Fluxogramas de processos são criados para facilitar o entendimento e padronização de procedimentos, gerar análises críticas, mas, também, porque eles ajudam a deixar claro para o colaborador o motivo pelo qual as atividades são feitas daquela forma. Dessa forma, um fluxograma não serve apenas para os programadores e sim para agrupar todas as pessoas que estão no processo, ou seja, ele é feito justamente para que haja um entendimento completo de todos. Nesse sentindo, você teria que pensar em uma lógica onde todos entendessem, melhorando assim sua lógica de programação e até mesmo a sua comunicação.

## Operadores de comparação
Como vimos na primeira aula de lógica, vimos que a linguagem *Python* possui alguns operadores lógicos que irão retornar um valor **True** ou **False**.

A sintaxe seria dessa forma
```
print(5 > 4)
print(5 >= 4)
print(5 <= 4)
print(10 == 4)
print(5 != 4)
```
Esses *print* irão retornar os nossos famosos tipos booleanos

Partindo desse conceito, podemos ver que podemos salvar até variáveis com esses operadores de comparação

````python
# Se atribuirmos uma comparação a uma variável, veremos que essa variável terá um valor guardado
x = 5 > 4
print(x)
````

Vemos que essa variável tem justamente o valor do resultado da comparação, que justamente é um valor Booleano.

## Condicionais
Como acabamos de ver nos operadores de comparação, as estruturas condicionais utilizam, também, esse efeito de comparação, onde verificam se o valor da condição é **True** ou **False**.

Dessa maneira, as estruturas condicionais fazem com que o código não seja executado de uma vez só, há uma interrupção no processo para verificar qual caminho o programa deve seguir.

Funciona da seguinte maneira:
```
if (condição):
  instruções
else:
  instruções
```

Quando utilizamos somente dessa forma, no bloco do *if* é necessário passar uma condição, ou seja, um efeito de comparação como vimos acima. Partindo desse pressuposto o *else*(senão) serve apenas para contradizer a condição do *if*.

Exemplo:

```
if 5 > 6:
  instruções
else:
  instruções
```

Dessa maneira, se 5 não for maior do que 6 ele vai cair no *else*, justamente contradizendo a instrução do *if*.



Certo, então caso a gente queira implementar outra condição, vamos utilizar uma estrutura chama *elif*, que em outras linguagens podem ser vistas como *else if*(senão se). Nesse caso, o *elif* precisa, também, de um efeito de comparação, ou seja, ele precisa de uma condição.

Exemplo:

```
if 5 > 5:
  (instruções)
elif 5 == 5:
  (instruções)
else:
  (instruções)
```

Nesse caso, o *elif* pode ser considerado uma estrutura intermediária dentro da seção de *if* e *else* e deve vir justamente quando temos mais uma verificação para ser feita.

Além desse caso, usando o efeito de comparação, temos o caso de verificação de variável, nesse caso, vamos usar o bloco condicional, para verificar se uma variável é verdadeira ou não.

Certo e quando essa variável seria "verdadeira"?

A condicional vai específicar a variável como verdadeira quando ela tiver algum conteúdo dentro dela, dessa forma

````python
nome = 'Felipe'

if nome:
  print('Essa variável é mesmo verdadeira')
````

````python
nome2 = ''

if nome2:
  print('Essa variável ainda é verdadeira')
````

Dessa forma, vimos que na célula de cima, onde a variável nome2 possui um valor vazio, o bloco *if*, não foi executado, justamente por não possuir um conteúdo ali dentro.

Ainda assim, temos um porém, não porque a variável não é verdadeira, que ela não tem um tipo, se verificarmos o tipo da variável nome2, será retornado um tipo, ou seja, não é que essa variável não esteja ocupando valor de memória dentro do nosso código, mas está ocupando um valor **False**.


````python
print(type(nome2))
````

Por último, temos a outra forma de utilizar um *if*, que é usando a palavra reservada *in*, essa palavra reservada vai servir para verificamos se há um determinado valor dentro de outro, nesse caso, podemos ver que ele é muito semelhante ao efeito de comparação e utilizamos ele da seguinte forma.

````python
# Veja, vamos pedir ao usuário para que ele digite o seu nome
nome_user = input('Digite o seu nome aí: ')

if 'Felipe' in nome_user:
  print('E aí Lipeta')
````

Certo, agora vamos detalhar o que está acontecendo aqui, estamos verificando se a *string*(campo de texto) está dentro da variável digitada pelo usuário, podemos usar esse *in* tanto com números como com textos.

## Agora vamos rever os nossos laços de repetição

#### O que são laços de repetição?

Os laços de repetição são recursos das linguagens de programação responsáveis por executar um bloco de código repetidamente através de determinadas condições especificas.

A linguagem Python possui dois tipos de estruturas de repetição que são
- *while*
- *for*


Primeiramente vamos rever o loop *for*. O comando *for* tem algumas forma para ser executados, primeiro podemos usar ele para percorrer estruturas, como *strings*(textos).

Exemplo:

```
for i in range 'Olá Mundo':
  print(i)
```

Neste caso, estamos usando o *for* para percorrer uma estrutura de texto utilizando a variável i, ou seja, cada vez que o *for* estiver sendo executado o i terá um valor diferente, primeiro o i terá o valor de O, depois l e assim por diante até finalizar o texto.

````python
# Exemplo:
for i in 'Olá mundo':
  print(i)
````

Outra forma de utilizarmos o comando *for* é de forma conjunta com a função *range()*, essa função permite especificarmos o início da sequência, o valor final obtido, e o passo, tudo isso através da seguinte sintáxe:
```
for i in range(0, 100, 1):
  print(i)
```

Nesse caso, se printarmos a variável i, veremos o valor de 0 até 99, pois, o valor do início é o primeiro e o valor do meio é o final da sequência, além disso temos o último valor que é 1, que informa qual será o incremento, ou seja, nesse caso será 1

Podemos **visualizar** esse *for* dessa forma:
```
for(i = 0; i < 100; i+1)
```
Nesse caso, é isso que a função range faz, ela atribui na primeira volta i a 0 e coloca uma condição, que nesse caso é para essa função ser executada enquanto o i é menor do que 100 e sempre somar o valor 1 mais o valor de i.

### Agora para finalizarmos, veremos o comando *while*, que é mais uma estrutra de repetição que temos na linguagem *Python*

Certo, como vimos na aula anterior, o comando *while* é outra forma de fazermos um loop em *Python*, porém, ele possui algumas diferenças do comando *for*, a primeira delas é a forma de usar, vamos utilizar o *while* **Normalmente**, quando não sabemos quando nosso laço irá terminar, ou seja, a estrutura será executada, **ENQUANTO** a condição é verdadeira.

Exemplo:
```
while (condição):
  instruções
```

Percebem que a sintáxe do *while* é muito semlhante ao do *if*? Então qual seria a diferença dele para o *if*?

A diferença dele para o *if* é que, o *while* irá ser executado enquanto a condição é verdadeira, já o *if* não, se a condição dele for verdadeira, ele vai executar o código apenas uma vez.

Então, usaremos essa estrutura quando é uma verificação que precisa ser executada de forma contínua.


````python
# Exemplo 1

# while 5 == 5:
#   print('Olá mundo')

# Se você executar o código acima verá que seu terminal vai imprimir Olá mundo diversas vezes sem parar
# Dessa forma, precisamos usar alguma forma para parar a execução do nosso código
# No primeiro momento vamos utilizar uma variável contadora, que é a famosa "flag"
````

````python
# Agora, criando uma variável e atribuindo ela a 0
cont = 0

# Agora faremos o while e vamos passar sua condição
while cont < 5:
  print(f'Hello World {cont}')
  # Agora precisamo incremetar nossa variável cont, se não ela sempre será falsa
  cont += 1
````

Veja que nosso programa parou, dessa vez, justamente por conta da condição do *while*, essa condição informa o mesmo **True** ou **False** que vimos em operadores.

Certo, mas não temos porque usarmos o *while* dessa forma, se já temos o comando *for*, nesse sentindo, a melhor maneira de usar o *while* é através do loop infinito.

Ou seja, quando não sabemos quando o loop vai parar.

O loop infinito segue essa sintáxe:

```
while True:
  intruções
```

De novo, como vimos acima, essas condições vão retornar sempre um **True** ou **False**, quando passamos já um **True** ao lado do *while*, já estámos informando que aquele condição é verdadeira, desse modo o loop ficará rodando de forma infinita, até pararmos a execução do nosso código.

````python
while True:
  print('Salve Mundo')
````

Porém, existe outras formas de pararmos esse loop, até porque podemos usar condicionais dentro desse loop e pararmos ele quando alguma dessas condicionais for verdadeira ou false, exemplo:

```
while True:
  nome = input('Digite seu nome: ')
  if nome in 'Felipe':
    
```
Vamos dizer que programa informa que toda vez que o usuário digitar Felipe o programa pare, utilizamos assim a nossa palavra reservada *break*, que informa que um loop deve ser parado.

````python

while True:
  nome = input('Digite seu nome: ')
  if nome in 'Felipe':
    break
````

Outro auxiliar que é comumente utilizado é o *continue*, informando que o loop deve continuar rodando, porém, aquele escopo específico deverá ser pulado

````python
num = 0
while num <= 5:
    if num == 2:
        print('Encontrei o 2')
        num += 1
        continue
    else:
        print(num)
        num += 1

    print('Não estou no IF')
````

Perceba, parece que ficou a mesma coisa que um *if* e *else* convencional certo?
Mas o *print* do "Estou abaixo do IF" não foi imprimido, mesmo ele não estando no escopo das condições, isso acontece porque assim que o programa encontra a palavra continue o loop volta para o início, não lingado pra nada ue esteja abaixo dele.
