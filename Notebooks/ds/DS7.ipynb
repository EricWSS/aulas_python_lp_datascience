````python
import pandas as pd
from matplotlib import pyplot as plt
````

###**plot()**

https://matplotlib.org/stable/api/pyplot_summary.html

````python
x = [*range(1, 5)]
y = [3, 5, 6, 3]

plt.plot(x, y)
plt.show()
````

O **plot()** possui dois paramêtros chamados de x e y, segundo a documentação, esses dois paramêtros vão adicionar a parte vertical e horizontal.

No fim utilizamos o *plt.show()* por ser uma boa prática, no Jupyter e no colab ele já "plota" o gráfico de  qualquer maneira, mas por uma boa prática, utilizamos o comando *show()*.

````python
plt.plot(x, y, label='Dados')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.show()
````

Podemos definir os labels e suas características através do xlabel e ylabel, eles estão, também, na documentação.

Além disso, podemos, também, dar títulos para os nossos gráficos

````python
plt.plot(x, y, label='Dados')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gráfico 1')
plt.show()
````

Além disso, podemos também mudar as escalas do eixo X e do eixo Y, utilizando o xticks e yticks.

````python
plt.plot(x, y, label='Dados')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gráfico 1')
plt.xticks([num for num in range(10) if num % 2 == 0])
plt.yticks([num for num in range(10) if num % 3 == 0])
plt.show()
````

Nesse exemplo acima modificamos os ticks, o eixo X ficou somente o 0, 2, 4, 6 e 8, e o eixo Y ficou o 0, 3, 6 e 9.

````python
plt.plot(x, y, label='Dados')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gráfico 1')
plt.xticks([num for num in range(10) if num % 2 == 0])
plt.yticks([num for num in range(10) if num % 3 == 0])
plt.show()
````

Além disso, o *plot()* possui outra propriedade chamada de linestyle, que vai definir o estilo da linha do nosso gráfico.

````python
plt.plot(x, y, label='Dados', linestyle='--')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gráfico 1')
plt.legend()
plt.show()
````

Podemos também, modificar a largura da linha do nosso gráfico, utilizando o **lw**

````python
plt.plot(x, y, label='Dados', linestyle='--', lw=4.0)
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gráfico 1')
plt.legend()
plt.show()
````

Vamos aumentar o tamanho da nossa *__figure__*

Nesse caso, é a figura inteira.

````python
plt.figure(figsize=(16, 6))
plt.plot(x, y, label='Dados', linestyle='--', lw=4.0)
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Gráfico 1')

plt.legend()

plt.show()
````

Vamos fazer mais exemplos:

````python
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']
valores = [105235, 107697, 110256, 109236, 108859, 109986]
````

````python
plt.plot(meses, valores)
plt.ylabel('Valores')
plt.xlabel('Meses')
plt.title('Vendas')
plt.show()
````

Nesse caso, acima, há um grande porém, como o gráfico foi gerado através do valor máximo e mínimo, pode ser que haja uma distorção visual, deixando parecer que os valores do gráfico estão muito diferentes um do outro e não é bem isso que acontece.

Assim, para evitar esse comportamento podemos mudar o valor do eixo vertical.

````python
plt.plot(meses, valores)
plt.ylabel('Valores')
plt.xlabel('Meses')
plt.title('Vendas')
plt.ylim(100_000, 120_000)
plt.show()
````

###Outras formas de plotarmos gráficos

Podemos utilizar os gráficos de *Scatter*, que seria o gráfico de disperção.

````python
plt.scatter(x, y)
plt.show()
````

Vamos mudar a cor dos pontos, com a propriedade color.

````python
plt.scatter(x, y, color='g')
plt.show()
````

````python
plt.scatter(x, y, color='g')
plt.show()
````

Dessa forma, nós temos praticamente os paramêtros que o gráfico de linha(plot), podemos mudar sua cor, o tamanho do gráfico e entre outras coisas presentes na documentação.

Além disso, temos o gráfico de barra, que chamamos de *__bar__*.

````python
plt.bar(x, y)
plt.show()
````

Temos vários tipos de gráficos, como por exemplo o de pizza, nesse caso, não veremos todos, porém, a documentação possui todos, com explicações e exemplos.

###Marcadores

Podemos alterar, também, os marcadores do gráfico, isso é uma outra forma de estilizar os gráficos para deixarmos em um padrão mais específico.

Dessa forma, temos os paramêtro *marker* que vai mudar os marcadores.

Abaixo temos a documentação específica para os markers.

https://matplotlib.org/stable/api/markers_api.html

````python
plt.scatter(x, y, color='g', marker='+')
plt.show()
````

##Figure

As figuras são como se fossem um concentrador de gráficos.

###Subplots

É a forma de fazer em uma mesma figura vários gráficos.

````python
x = [*range(1, 5)]
y = [3, 5, 6, 3]


fig = plt.figure(figsize=(20, 12))
fig.suptitle('Título da figura geral')
plt.show(fig)
````

Nesse caso, temos 0 quadrantes e precisamos criar eles.

````python
x = [*range(1, 5)]
y = [3, 5, 6, 3]


fig = plt.figure(figsize=(20, 3))
fig.suptitle('Título da figura')

fig.add_subplot(131)
plt.plot(x, y, label='Dados')
plt.legend()
plt.title('Primeiro gráfico')

plt.show(fig)
````

O subplot(131) significa que a figura possui uma linha, três colunas e que o gráfico que estamos plotando está na primeira posição.

````python
x = [*range(1, 5)]
y = [3, 5, 6, 3]


fig = plt.figure(figsize=(20, 3))
fig.suptitle('Título da figura')

fig.add_subplot(131)
plt.plot(x, y, label='Dados')
plt.legend()
plt.title('Primeiro gráfico')


fig.add_subplot(133)
plt.bar(x, y)
plt.title('Gráfico de barra')


fig.add_subplot(132)
plt.scatter(x, y)
plt.title('Gráfico de disperção')

plt.show(fig)
````
